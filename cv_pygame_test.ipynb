```python
def cvimage_to_pygame(image):
    """Convert cvimage into a pygame image"""
    return pygame.image.frombuffer(image.tostring(), image.shape[1::-1],
                                   "RGB")
```

```python
pygame.init()
```

```python
class Capture(object):
    def __init__(self, vc):
        self.size = (640,480)
        # create a display surface. standard pygame stuff
        self.display = pygame.display.set_mode(self.size, 0)
    
    
        self.vc = vc
        retval, frame = self.vc.read()
        #self.image = image
        self.snapshot = cvimage_to_pygame(frame)
        # create a surface to capture to.  for performance purposes
        # bit depth is the same as that of the display surface.
        #self.snapshot = pygame.surface.Surface(self.size, 0, self.display)
        self.myfont = pygame.font.SysFont("comicsansms", 36, bold=True)

    def get_and_flip(self):
        # if you don't want to tie the framerate to the camera, you can check 
        # if the camera has an image ready.  note that while this works
        # on most cameracams, some will never return true.
        #if self.cam.query_image():
        #    self.snapshot = self.cam.get_image(self.snapshot)

        retval, frame = self.vc.read()
        rgbframe = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        #self.image = image
        self.snapshot = cvimage_to_pygame(rgbframe)
        '''
        pygame.draw.circle(self.snapshot,
                           (255,255,255),
                           (320, 240),
                           20)
                           '''
        txtsurf = self.myfont.render('forseti2 vlay', True, (255,0,0))
        
        # blit it to the display surface.  simple!
        self.display.blit(self.snapshot, (0,0))
        self.display.blit(txtsurf, (100, 10))
        pygame.display.flip()

    def main(self):
        going = True
        while going:
            events = pygame.event.get()
            self.get_and_flip()
```

```python
cap = Capture(vc)
```

```python
cap.main()
```
